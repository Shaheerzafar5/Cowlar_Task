```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import StandardScaler

# Read the dataset
movies_data = pd.read_csv("movies_dataset.csv")

X = movies_data.iloc[:, 1:].values

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test = train_test_split(X_scaled, test_size=0.2, random_state=42)

cos_sim_matrix = cosine_similarity(X_test, X_train)

# Recommendation Generation
def recommend_movie(input_movie):
    input_index = movies_data.index[movies_data['title'] == input_movie].tolist()[0]

    similar_movie_index = cos_sim_matrix[input_index].argmax()

    recommended_movie = movies_data.iloc[similar_movie_index]['title']

    return recommended_movie

# Example usage:
input_movie = input("Enter Movie title: ")
recommended_movie = recommend_movie(input_movie)
print("Recommended movie:", recommended_movie)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.neighbors import NearestNeighbors
from sklearn.preprocessing import StandardScaler

movies_data = pd.read_csv("movies_dataset.csv")

X = movies_data.iloc[:, 1:].values
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_train, X_test = train_test_split(X_scaled, test_size=0.2, random_state=42)
cos_sim_matrix = cosine_similarity(X_test, X_train)

k = 5  # Number of neighbors to consider
nn_model = NearestNeighbors(n_neighbors=k, metric='cosine')
nn_model.fit(X_scaled)

# Step 5: Recommendation Generation
def recommend_movie(input_movie):
    input_index = movies_data.index[movies_data['title'] == input_movie].tolist()[0]
    similar_movie_index_cosine = cos_sim_matrix[input_index].argmax()
    _, nn_indices = nn_model.kneighbors([X_scaled[input_index]])
    recommended_movie_nn = movies_data.iloc[nn_indices[0][1]]['title']  # We skip the first index which is the input movie
    return recommended_movie_nn

# Example usage:
input_movie = input("Enter movie title: ")
recommended_movie_nn = recommend_movie(input_movie)
print("Recommended movie using Nearest Neighbors:", recommended_movie_nn)
```
